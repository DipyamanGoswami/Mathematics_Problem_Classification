```python
from sentence_transformers import SentenceTransformer
model= SentenceTransformer('all-mpnet-base-v2')
```

```python
import spacy
nlp= spacy.load("en_core_web_sm")
```

```python
stopwords= nlp.Defaults.stop_words
stopwords
```

```python
import string
punc= string.punctuation
list(punc)
```

```python
def tokenize(sentence):
  doc= nlp(sentence)

  # lemmatizing each token to convert it into base form and then converting it to lower case and removing any whitespaces
  tokens= [w.lemma_.lower().strip() for w in doc]

  #print(tokens)

  # removing any stop words and punctuations present in the sentences
  new_tokens= [x for x in tokens if x not in stopwords and x not in punc]

  sentences= " ".join(new_tokens)
  return sentences
```

```python
tokenize("My name is Dipyaman and I love programming") # A testcase
```

```python
tokenized_data= train_data['problem'].apply(tokenize)
#tokenized_data
```

```python
embeddings= tokenized_data.apply(model.encode)
#embeddings
```

```python
train_data['tokenized_data']= tokenized_data
train_data['embeddings']= embeddings
#train_data
```

```python
x= train_data['embeddings'].to_list()
y= train_data['category'].to_list()
```

```python
test_data= pd.read_csv("test.csv")
#test_data
```

```python
import re
regex= re.compile(r'<[^>]+>')

def remove_tags(sentence):
  return regex.sub('', sentence)
```

```python
processed_test_data= test_data['problem'].apply(remove_tags)
#processed_test_data
```

```python
tokenized_test_data= processed_test_data.apply(tokenize)
#tokenized_test_data
embedded_test_data= tokenized_test_data.apply(model.encode)
#embedded_test_data
```

```python
test_data['processed_test_data']= processed_test_data
test_data['tokenized_test_data']= tokenized_test_data
test_data['embedded_test_data']= embedded_test_data
#test_data
```

```python
x_to_pred= test_data['embedded_test_data'].to_list()
#x_test
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

mdl= KNeighborsClassifier()

x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.2, random_state=8)

mdl.fit(x_train, y_train)
```

```python
y_pred= mdl.predict(x_test)
y_pred
```

```python
accuracy= accuracy_score(y_pred, y_test)
accuracy
```
